```python
exp_numbers = []
for i in np.arange(0, 11):
    exp_numbers.append(i**2)
print(exp_numbers)
```

```python
for i, exp_number in enumerate(exp_numbers):
    print(i, exp_number)
```

```python
div_by_5 = False
i = 0

while not div_by_5:
    num = exp_numbers[i]
    
    if num == 0:
        i = i + 1
        continue
    
    if (num%5 == 0):
        print(num)
        div_by_5 = True
    else:
        i = i + 1
        if i >= len(exp_numbers):
            div_by_5 = True
            break
```

```python
def celsius_to_fahrenheit(temp_c):
    """This function will take a temperature in Celsius and convert it to Fahrenheit.

    Arguments:
        temp_c: Temperature in Celsius

    Returns:
        temp_f: Temperature in Fahrenheit
    """
    temp_f = temp_c * (9 / 5) + 32
    return temp_f
```

```python
new_temp_f = celsius_to_fahrenheit(40)
print(new_temp_f)
```

```python
def convert_temp(temp_in=-999, convert_to="C"):
    """This function will take a temperature in Celsius and convert to Fahrenheit.

    Arguments:
        temp_in : Temperature in Celsius or Farenheit
        convert_to: conversion to 

    Returns:
        temp_out: Temperature in Fahrenheit, Celsius, or -999 if nothing is calculated
    """
    
    if convert_to == "F":
        temp_out = temp_in * (9 / 5) + 32
    elif convert_to == "C":
        temp_out = (temp_in - 32) * (5 / 9)
    else:
        print("Unrecognized conversion")
        temp_out = -999
    return temp_out
```

```python
def daily_precip(*args):
    """This function calculates the total rainfall of any number of arguments
    
    Arguments:
        hourly rain totals
    Returns:
        Total precipitation
    """
    result = 0
    for arg in args:
        result = result + arg
    return result
```

```python
daily_precip(12, 20, 33)
```

```python
def daily_precip_list(args):
    result = 0
    for arg in args:
        result = result + arg
    return result
```

```python
hourly_precip = [0, 12, 20, 33]
daily_precip_list(hourly_precip)
```

## 4.11 Classes


```python
class OceanReading:
    def __init__(self, station, lat, lon, sst_F, wave_height_ft, \
            wave_direction_deg, date, time_utc):
        self.station = station
        self.lat = lat
        self.lon = lon
        self.sst = sst_F
        self.wave_height = wave_height_ft
        self.wave_direction = wave_direction_deg
        self.date = date
        self.time_utc = time_utc

    def __repr__(self):
        return (f"OceanReading(station='{self.station}', "
                f"lat={self.lat}, lon={self.lon}, "
                f"sst_F={self.sst}, date='{self.date}')")

    def to_celsius(self):
        return round((self.sst - 32) * 5/9, 1)

    def to_meters(self):
        return round(self.wave_height * 0.3048, 1)

    def speed_restrictions(self):
        dt_date = datetime.strptime(self.date, "%Y-%m-%d")
        return dt_date >= datetime(2026, 11, 1) and \
            dt_date <= datetime(2027, 4, 30)
```

```python
reading = OceanReading(station="Station_44099", \
        lat=36.915, lon=-75.722, sst_F=59.2, \
        wave_height_ft=3.9, wave_direction_deg=69, \
        date='2026-04-25', time_utc='18:56:00')
```

```python
print(reading)
```

```python
print("SST (Celsius):", reading.to_celsius())
print("Wave height (meters):", reading.to_meters())
print("Speed restrictions?", reading.speed_restrictions())
```

```python
reading.sst = 60.0
print("Updated sst:", reading.sst)
```
